```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"/Users/timothypark/Documents/portfolios/timpark99.github.io/Pandas Visualizations/Ice Cream Ratings.csv")
df = df.set_index('Date')
df
```

```python
# different styles you can use to change the visualization

print(plt.style.available)
plt.style.use('Solarize_Light2')
```

```python
# type shift then tab in plot to get more information
# it will default to a line plot
# if we want the plots of individual columns we type subplots=True

# df.plot(kind='line', subplots=True)

df.plot(kind='line', title='Ice Cream Ratings', xlabel='Daily Ratings', ylabel='Scores')
```

```python
# create a bar chart and stack the column values

df.plot(kind='bar', stacked=True)
```

```python
# create a bar chart specifically for one column

df['Flavor Rating'].plot(kind='bar', stacked=True)
```

```python
# create a horizontal bar chart that is stacked

df.plot.barh(stacked=True)
```

```python
# create a scatter plot.  We need to input the x and the y
# s changes the size of the point
# c changes the color

df.plot.scatter(x='Texture Rating', y='Overall Rating', s=100, c='Red')
```

```python
# create a histogram - shows distribution of variables

df.plot.hist(bins=20)
```

```python
# create box plot that shows from bottom to top: minimum value, 25th percentile, 50th percent, 75th percent and then max value
# can see which way the data skews towards

df.boxplot()
```

```python
# create area plot.  Similar to a line plot
# figsize can make it bigger or smaller

df.plot.area(figsize=(10,5))
```

```python
# create a pie chart
# you need to input arguments otherwise it'll produce an error

df.plot.pie(y='Flavor Rating', figsize=(10,10))
```
