# Indexing

```python
# index is an object that stores the axis labels for all pandas objects
# index in a dataframe is customizable
# index helps with search and filter

import pandas as pd

# show only 20 rows
pd.set_option('display.max.rows', 20)
```

```python
# use encoding='latin-1' since it shows error of UTF-8

df = pd.read_csv(r"/Users/timothypark/Documents/portfolios/timpark99.github.io/Indexes in Pandas/Popular_Spotify_Songs.csv", encoding='latin-1')

df
```

```python
# index is a number or a label for each row
# indexes should be unique but it doesn't have to be a number
# track name can be the index since there will be no duplicate track name
# index_col will change the index to track_name
# index indicated by bold letters and on the far left

df = pd.read_csv(r"/Users/timothypark/Documents/portfolios/timpark99.github.io/Indexes in Pandas/Popular_Spotify_Songs.csv", encoding='latin-1', index_col='track_name')

df
```

```python
# reset the index
# inplace=True allows us not to assign to another variable

df.reset_index(inplace=True)
```

```python
# index was reset

df
```

```python
# another way to change index but it won't save

df.set_index('track_name')
```

```python
# change index and save it using inplace=True.  We don't have to save it to another variable

df.set_index('track_name', inplace=True)

df
```

```python
# loc stands for location and iloc stands for integer location

df.loc['LALA']
```

```python
# using iloc instead of loc
# since the song LALA is in the index 1 position we get the same result

df.iloc[1]
```

```python
# reset the index to original df

df.reset_index(inplace=True)
```

```python
# multi indexing

df.set_index(['released_year','track_name'], inplace=True)
```

```python
# group together the tracks released by year

df.sort_index()
```

```python
# it can be sorted in descending order

df.sort_index(ascending=False)
```

```python
# it can be sorted as a list
# the release date will be descending and the track name will be ascending

df.sort_index(ascending=[False, True])
```

```python
# this won't work because it's in axis 1 and not axis 0
# loc will only look in axis 0

df.loc['AMARGURA']
```

```python
# here we can perform loc on the year
# here are all the songs in 2023

df.loc[2023]
```

```python
# here we can perform loc on a list
# here is the data for the song AMARGURA

df.loc[2023, 'AMARGURA']
```

```python
# loc is able to specify multi indexes
# iloc cannot do this and will go based off of the initial index

df.iloc[1]
```
