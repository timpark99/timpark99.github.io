 <div class="alert alert-block alert-success"><strong> Problem 1:</strong><p>
A manufacturer sells engine belts for 14 dollars per belt. The fixed costs are 720 dollars, and the variable costs are 5 dollars per unit.

a) Write the cost function C(x). 
 
  

b) Write the revenue function R(x). 
 
  

c) Write the profit function P(x). 
 
  

d) The number of engine belts that the manufacture must sell to break even is 
 belts.</div>

```python
# Given values
fixed_cost = 720
variable_cost = 5
price = 14

# Cost function
def C(x):
    return fixed_cost + variable_cost * x

# Revenue function
def R(x):
    return price * x

# Profit function
def P(x):
    return R(x) - C(x)

# Break-even calculation
# occurs when R(x) = C(x) - solve for x
break_even = fixed_cost / (price - variable_cost)

print("Cost function: C(x) = 720 + 5x")
print("Revenue function: R(x) = 14x")
print("Profit function:  P(x) = 9x - 720")
print("Break-even point:", break_even)
```

<div class="alert alert-block alert-success"><strong> Problem 2:</strong><p>
Find the inverse of the invertible matrix  
\[
B = \begin{bmatrix}
1 & -1 & -3 \\
-10 & 11 & 28 \\
2 & -2 & -5
\end{bmatrix}.
\]

Compute \( B^{-1} \) using Python.</div>

```python
import sympy as sp

# Define the matrix B
B = sp.Matrix([
    [1, -1, -3],
    [-10, 11, 28],
    [2, -2, -5]
])

# Compute the inverse
B_inv = B.inv()

print("Matrix B:")
print(B)
print("\nInverse of B:")
print(B_inv)
```

<div class="alert alert-block alert-success"><strong> Problem 3:</strong><p>
### Solve the System of Equations

Solve the following system of equations:

\[
\begin{cases}
x + 2y + 6z = 13 \\
-x - 5y - 4z = 16 \\
3x - y - 3z = 4
\end{cases}
\]

Use Python (e.g., with `sympy`) to find the solution \((x, y, z)\).
</div>

```python
import sympy as sp

# Define variables
x, y, z = sp.symbols('x y z')

# Define the system of equations:
# x + 2y + 6z = 13
# -x - 5y - 4z = 16
# 3x - y - 3z = 4
eq1 = sp.Eq(x + 2*y + 6*z, 13)
eq2 = sp.Eq(-x - 5*y - 4*z, 16)
eq3 = sp.Eq(3*x - y - 3*z, 4)

# Solve the system
solution = sp.solve((eq1, eq2, eq3), (x, y, z))
print(solution)
```

<div class="alert alert-block alert-success"><strong> Problem 4:</strong><p>
## Museum Field Trip Problem

Admission to a museum is 30 dollars for adults (teachers) and 20 dollars for students.  
A total of 234 people attended, and $4750 was collected.

Let  
- **s** = number of students  
- **t** = number of teachers  

### (a) Total number of people
\[
s + t = 234
\]

### (b) Total money collected
\[
20s + 30t = 4750
\]

</div>

```python
import sympy as sp

# Define variables
s, t = sp.symbols('s t')

# Equations:
# Total people
eq1 = sp.Eq(s + t, 234)

# Money collected: $20 per student, $30 per teacher
eq2 = sp.Eq(20*s + 30*t, 4750)

solution = sp.solve((eq1, eq2), (s, t))
solution
```

## Problem
Given the matrix  
\[
A = 
\begin{bmatrix}
-33 & -96 & 70 \\
0 & 1 & 0 \\
-15 & -42 & 32
\end{bmatrix},
\]
determine whether **â€“3 is an eigenvalue** of \(A\).

A number \(\lambda\) is an eigenvalue of \(A\) if  
\[
\det(A - \lambda I) = 0.
\]


</div>

```python

import sympy as sp

A = sp.Matrix([
    [-33, -96, 70],
    [  0,   1,  0],
    [-15, -42, 32]
])

lam = -3
det_val = (A - lam*sp.eye(3)).det()
print(det_val)
```

<div class="alert alert-block alert-success"><strong> Problem 8:</strong><p>
Giselle manages a store that sells speakers. Each **Thumper** takes 4 hours to create the plastic pieces, 2 hours to create the electronics, and 13 hours of labor to assemble the final product. Each **XL2000** requires 2 hours to create the plastic pieces, 4 hours to create the electronics, and 10 hours of labor to assemble the final product. 

The factory has 320 hours to create the plastic pieces, 380 hours to create the electronics, and 1110 hours of labor to assemble the final product each week. If each Thumper generates 13 dollars in revenue, and each XL2000 generates 19 dollars, how many of each of the speakers should Giselle have the store sell each week to maximize revenue?

- Number of Thumper speakers to sell:  
- Number of XL2000 speakers to sell:  
- Maximum revenue: \$

</div>

```python
import numpy as np
from scipy.optimize import linprog

# We want to maximize 13T + 19X,
# linprog minimizes, so minimize -13T - 19X
c = np.array([-13, -19])

# Inequality constraints A @ [T, X] <= b
A = np.array([
    [4, 2],    # plastic
    [2, 4],    # electronics
    [13, 10]   # labor
])
b = np.array([320, 380, 1110])

# Bounds: T >= 0, X >= 0
bounds = (0, None)

result = linprog(c, A_ub=A, b_ub=b, bounds=bounds)

print("Optimal T, X:", result.x)
print("Maximum revenue:", -result.fun)
```
