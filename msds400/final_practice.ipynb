 <div class="alert alert-block alert-success"><strong> Problem 1:</strong><p>
A manufacturer sells engine belts for 14 dollars per belt. The fixed costs are 720 dollars, and the variable costs are 5 dollars per unit.

a) Write the cost function C(x). 
 
  

b) Write the revenue function R(x). 
 
  

c) Write the profit function P(x). 
 
  

d) The number of engine belts that the manufacture must sell to break even is 
 belts.</div>

```python
# Given values
fixed_cost = 720
variable_cost = 5
price = 14

# Cost function
def C(x):
    return fixed_cost + variable_cost * x

# Revenue function
def R(x):
    return price * x

# Profit function
def P(x):
    return R(x) - C(x)

# Break-even calculation
# occurs when R(x) = C(x) - solve for x
break_even = fixed_cost / (price - variable_cost)

print("Cost function: C(x) = 720 + 5x")
print("Revenue function: R(x) = 14x")
print("Profit function:  P(x) = 9x - 720")
print("Break-even point:", break_even)
```

<div class="alert alert-block alert-success"><strong> Problem 3:</strong><p>
### Solve the System of Equations

Solve the following system of equations:

\[
\begin{cases}
x + 2y + 6z = 13 \\
-x - 5y - 4z = 16 \\
3x - y - 3z = 4
\end{cases}
\]

Use Python (e.g., with `sympy`) to find the solution \((x, y, z)\).
</div>

```python
import sympy as sp

# Define variables
x, y, z = sp.symbols('x y z')

# Define the system of equations:
# x + 2y + 6z = 13
# -x - 5y - 4z = 16
# 3x - y - 3z = 4
eq1 = sp.Eq(x + 2*y + 6*z, 13)
eq2 = sp.Eq(-x - 5*y - 4*z, 16)
eq3 = sp.Eq(3*x - y - 3*z, 4)

# Solve the system
solution = sp.solve((eq1, eq2, eq3), (x, y, z))
print(solution)
```

<div class="alert alert-block alert-success"><strong> Problem 8:</strong><p>
Giselle manages a store that sells speakers. Each **Thumper** takes 4 hours to create the plastic pieces, 2 hours to create the electronics, and 13 hours of labor to assemble the final product. Each **XL2000** requires 2 hours to create the plastic pieces, 4 hours to create the electronics, and 10 hours of labor to assemble the final product. 

The factory has 320 hours to create the plastic pieces, 380 hours to create the electronics, and 1110 hours of labor to assemble the final product each week. If each Thumper generates 13 dollars in revenue, and each XL2000 generates 19 dollars, how many of each of the speakers should Giselle have the store sell each week to maximize revenue?

- Number of Thumper speakers to sell:  
- Number of XL2000 speakers to sell:  
- Maximum revenue: \$

</div>

```python
import numpy as np
from scipy.optimize import linprog

# We want to maximize 13T + 19X,
# linprog minimizes, so minimize -13T - 19X
c = np.array([-13, -19])

# Inequality constraints A @ [T, X] <= b
A = np.array([
    [4, 2],    # plastic
    [2, 4],    # electronics
    [13, 10]   # labor
])
b = np.array([320, 380, 1110])

# Bounds: T >= 0, X >= 0
bounds = (0, None)

result = linprog(c, A_ub=A, b_ub=b, bounds=bounds)

print("Optimal T, X:", result.x)
print("Maximum revenue:", -result.fun)
```

## ðŸ“˜ Optimization Problem: Jewelry-Making Kits

The total cost to produce \(x\) large jewelry-making kits and \(y\) small ones is given by:

\[
C(x, y) = 2x^2 + 6y^2 + 4xy + 10
\]

A total of **10 kits** must be made, so the constraint is:

\[
x + y = 10
\]

### **Tasks**
1. Determine how many **large kits** should be produced.  
2. Determine how many **small kits** should be produced.  
3. Find the **minimum total cost**.

---
</div>

```python
import sympy as sp

# Variables
x = sp.symbols('x', real=True)
y = 10 - x   # constraint x + y = 10

# Cost function
C = 2*x**2 + 6*y**2 + 4*x*y + 10

# Derivative of cost w.r.t x
dC_dx = sp.diff(C, x)

# Solve for critical point
crit = sp.solve(sp.Eq(dC_dx, 0), x)

# Compute optimal x, y, and cost
x_opt = crit[0]
y_opt = 10 - x_opt
C_opt = C.subs(x, x_opt)

x_opt, y_opt, C_opt
```

## ðŸ“˜ Marginal Cost Problem

The cost (in hundreds of dollars) to produce \(x\) units of a product is

\[
C(x) = \frac{6x - 7}{4x^2 + x}.
\]

Find the **marginal cost** when producing 8 units.  
Round your answer to the nearest cent.

Remember:  
- The derivative \(C'(x)\) gives marginal cost **in hundreds of dollars**,  
- So multiply the derivative by 100 to convert to dollars.


</div>

```python
import sympy as sp

# Define variable
x = sp.symbols('x', positive=True)

# Cost function (in hundreds of dollars)
C = (6*x - 7) / (4*x**2 + x)

# Derivative = marginal cost
dC = sp.diff(C, x)

# Evaluate marginal cost at x = 8 (convert from hundreds of dollars to dollars)
marginal_cost = float(dC.subs(x, 8) * 100)

marginal_cost
```

## ðŸ“˜ Instantaneous Rate of Change of Population

The projected population (in thousands) of a certain ethnic group is given by the function:

\[
p(t) = 35.11 e^{0.54t}
\]

where  
- \( t = 0 \) corresponds to the year **2010**,  
- and 0 <= t <= 50).

### **Question**
Find the **instantaneous rate of change** of the population in the year **2020**.  
Round your answer to the nearest whole number.

### Notes:
- Since \(t = 0\) is 2010, the year 2020 corresponds to \(t = 10\).
- The function outputs population **in thousands**, so the derivative must be multiplied by 1000 to convert to **individuals per year**.

</div>

```python
import sympy as sp

# Define variable
t = sp.symbols('t', real=True)

# Population model (in thousands)
p = 35.11 * sp.exp(0.54 * t)

# Derivative of the population
dp = sp.diff(p, t)

# Evaluate the instantaneous rate of change at t = 10 (year 2020)
rate_2020 = float(dp.subs(t, 10) * 1000)  # convert thousands â†’ individuals

rate_2020
```

## ðŸŒ³ Minimum Cost Spanning Tree using Kruskalâ€™s Algorithm

We are given the following weighted graph with vertices:

- Left side: A, B, C  
- Right side: D, E, F  

Each edge has a cost (weight). The edges and weights interpreted from the diagram are:

- Aâ€“D : 2  
- Aâ€“E : 7
- Aâ€“F : 14  
- Bâ€“D : 11  
- Bâ€“E : 5  
- Bâ€“F : 4  
- Câ€“D : 6  
- Câ€“E : 3  
- Câ€“F : 12  

Using **Kruskalâ€™s Algorithm**, we want to determine which edges belong to the **minimum cost spanning tree (MST)**.

### Goal  
Find the set of edges in the MST.

Below is the Python code that computes the MST using NetworkX.

</div>

```python
import networkx as nx

# Create graph
G = nx.Graph()

# Edges interpreted from the diagram (node1, node2, weight)
edges = [
    ("A", "D", 2),
    ("A", "E", 7),
    ("A", "F", 14),
    ("B", "D", 11),
    ("B", "E", 5),
    ("B", "F", 4),
    ("C", "D", 6),
    ("C", "E", 3),
    ("C", "F", 12)
]

# Add edges to graph
G.add_weighted_edges_from(edges)

# Compute MST using Kruskal (NetworkX uses Kruskal by default)
mst = nx.minimum_spanning_tree(G)

# Display MST edges
mst_edges = list(mst.edges(data=True))
mst_edges
```
