<div class="alert alert-block alert-success"><strong> Problem 1:</strong><p>
Is it possible for a planar graph to have 6 vertices, 9 edges and 4 faces?</div>

```python
# Euler's formula for planar graphs is:
#   V - E + F = 2
#
# where
#   V= number of verticies
#   E= number of edges
#   F= number of faces

# Parameters of problem:
V = 6  # Number of vertices
E = 9  # Number of edges
F = 4  # Number of faces

# Euler's formula
euler_result = V - E + F
print ("Euler Result: ", euler_result)

# If the euler result is 2 , then it is possible for a planar graph.
# Otherwise, it is not.
```

<div class="alert alert-block alert-success"><strong> Problem 2:</strong><p>
The graph G has 5 vertices with degrees 2, 3, 3, 4, 4. <ul><li>How many edges does G have?</li><li>Could G be planar?</li><li>If it could be planar, how many faces would it have?</li></ul></div>

```python
# Parameters of problem:
degrees = [2, 3, 3, 4, 4]  # degrees of vertices

# The number of edges in a graph can be calculated using the degrees of the vertices. 
# Since each edge is connected to two vertices, the sum of the degrees of all vertices is twice the number of edges. 

# So, the number of edges E can be calculated as:
E = sum(degrees) // 2
print ("Number of edges: E= ", E)

V = len(degrees)           # number of vertices
print ("Number of vertices: V=", V)

# Solve Euler's formula for F, assuming that it possible for it to be planar
#    F = 2 - V + E
F = 2 - V + E
print ("Number of faces: F= ", F)

# If it's possible to be a planar graph or not, we check by Euler's formula: V - E + F = 2
# where V= number of vertices, E= number of edges, F= number of faces

print ("")
euler_result = V - E + F
print ("Euler result: V - E + F = ", euler_result)

# If the euler result is 2 , then it is possible for a planar graph.
# Otherwise, it is not.
```

<div class="alert alert-block alert-success"><strong>Problem 6:</strong><p>Which of the graphs is NOT a tree?<br><em>(Only the first two graphs (a) and (b) are computed for you in this example.)</em> </div>

```python
import networkx as nx

G_a= nx.Graph()
G_a.add_nodes_from(['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h'])
G_a.add_edge('a','d') # 1
G_a.add_edge('b','e') # 2
G_a.add_edge('b','h') # 3
G_a.add_edge('c','d') # 4
G_a.add_edge('e','g') # 5
G_a.add_edge('d','f') # 6                
G_a.add_edge('f','g') # 7
G_a.add_edge('f','h') # 8
# Double-checking the diagram: there are 8 edges!
# Also, make sure that you're not double-counting any edges.

print ("Graph (a) is a tree:" , nx.is_tree(G_a))

G_b= nx.Graph()
G_b.add_nodes_from(['a', 'b', 'c', 'd', 'e'])
G_b.add_edge('a','b') # 1
G_b.add_edge('a','c') # 1
G_b.add_edge('a','d') # 3
G_b.add_edge('d','e') # 4
# Double-checking the diagram: there are 4 edges!

print ("Graph (b) is a tree:" , nx.is_tree(G_b))
```

<div class="alert alert-block alert-success"><strong>
    Problem 8:</strong> 
<ul>
<li>(a) How many internal vertices and leaves does a full 6-ary tree have with 37 vertices?</li>
<li>(b) How many vertices and leaves does a full 5-ary tree have with 7 internal vertices?</li>
<li>(c) How many vertices and internal vertices does a full 7-ary tree have with 67 leaves?</li>
</ul></div>

```python
def full_tree_properties(m, v):
    i = (v - 1) // m  # Theorem (i): i = (n - 1) / m
    l = ((m - 1) * v + 1) // m  # Theorem (i): l = ((m - 1)n + 1) / m
    return i, l


# Full 5-ary tree with 41 vertices
m = 5
v = 41
i, l = full_tree_properties(m, v)
print(f"A full {m}-ary tree with {v} vertices has {i} internal vertices and {l} leaves.")


# Full 3-ary tree with 10 internal vertices
m = 3
i = 10
v = m * i + 1 # Theorem 3:  n = m*i + 1
i, l = full_tree_properties(m, v)
print(f"A full {m}-ary tree with {i} internal vertices has {v} vertices and {l} leaves.")


# Full 6-ary tree with 31 leaves
m = 6
l = 31
v = (m*l)//(m-1) # Theorem (iii):  n = (ml-1) / (m-1)
i = (l-1)//(m-1) # Theorem (iii):  i =  (l-1) / (m-1)
print(f"A full {m}-ary tree with {l} leaves has {v} vertices and {i} internal vertices.")
```

<div class="alert alert-block alert-success"><strong> Problem 10:</strong> Build a binary search tree for the following words, sorted alphabetically: gaffe, rebellion, fool, elaborate, spread, joke, freedom, stroke, guideline, and aware. Add words to this tree in the order in which they are listed.<br>
<ul>    
<li>(a) How many comparisons are needed to locate spread in this tree, starting from the top?</li>

<li>(b) How many comparisons are needed to add the word thorough to this tree?</li>

<li>(c) What is the parent of the node labeled elaborate?<br>
gaffe, rebellion, fool, elaborate, spread, joke, freedom, stroke, guideline, aware</li>

<li>(d) What are the children of the node labeled fool?<br>
gaffe, rebellion, fool, elaborate, spread, joke, freedom, stroke, guideline, aware</li>
</ul>
</div>

```python
# This solution uses recursion, which we may not have talked about before.

def add_word(tree, word):
    # If the tree is empty, add the word as the root
    if not tree:
        print ("Inserting ", word, "\n")
        tree.append(word)
        tree.append([])  # left subtree
        tree.append([])  # right subtree
    else:
        print ("Comparing ", word, " to ", tree[0])
        # If the word is less than the root, add it to the left subtree
        if word < tree[0]:
            print ("  Going to the left of ", tree[0])
            add_word(tree[1], word)
        # If the word is greater than the root, add it to the right subtree
        elif word > tree[0]:
            print ("  Going to the right of ", tree[0])
            add_word(tree[2], word)

            
def print_tree(tree, indent=0):
    if tree:
        print_tree(tree[2], indent + 4)  # Print right subtree
        print(' ' * indent + tree[0])  # Print root
        print_tree(tree[1], indent + 4)  # Print left subtree

        
words = ["gaffe", "rebellion", "fool", "elaborate", "spread", "joke", "freedom", "stroke", "guideline", "aware"]
tree = []

print ("\nHow the comparisons and inserts of the binary tree are handled:\n")
for word in words:
    add_word(tree, word)

print ("\nTree:\n")
print ("RIGHT SIDE OF TREE\n")
print_tree(tree)
print ("\nLEFT SIDE OF TREE")
```

```python
# Looking at the sideways printed version of the tree above...

# (a) How many comparisons are needed to locate "spread" in this tree, starting from the top?
# ANSWER: 2; gaffe, rebellion --> "spread"

# (b) How many comparisons are needed to add the word "thorough" to this tree?
# ANSWER: 4; gaffe, rebellion, spread, stroke --> "thorough"

# (c) What is the parent of the node labeled "elaborate"?
# gaffe, rebellion, fool, elaborate, spread, joke, freedom, stroke, guideline, aware
# ANSWER: "fool"
    
# (d) What are the children of the node labeled "fool"?
# gaffe, rebellion, fool, elaborate, spread, joke, freedom, stroke, guideline, aware
# ANSWER: "freedom" and "elaborate"
```
