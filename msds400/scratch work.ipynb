```python
from sympy import symbols, Eq, solve

# Define variables
x, y, z = symbols('x y z')

# Equations
eq1 = Eq(1*x + 6*y + -2*z, -61)                 
eq2 = Eq(0*x + 3*y + 2*z, -11) 
eq3 = Eq(1*x + -2*y + -1*z, 0)

# Solve system
solution = solve((eq1, eq2, eq3), (x, y, z))
print(solution)
```

```python
import numpy as np

# Example system of equations:
# a1*x + b1*y + c1*z = d1
# a2*x + b2*y + c2*z = d2
# a3*x + b3*y + c3*z = d3

# Define coefficient matrix A
A = np.array([
    [5, -5, 5],   # coefficients of eqn 1
    [0, 1, 2],    # coefficients of eqn 2
    [1, -2, -1]   # coefficients of eqn 3
])

# Define constants vector B
B = np.array([-55, -11, 0])

# Solve for [x, y, z]
solution = np.linalg.solve(A, B)

x, y, z = solution
print(f"x = {x}, y = {y}, z = {z}")
```
