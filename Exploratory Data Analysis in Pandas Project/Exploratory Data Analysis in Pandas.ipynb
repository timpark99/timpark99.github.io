# EDA in Pandas

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"/Users/timothypark/world_population.csv")
df
```

```python
pd.set_option('display.float_format', lambda x: '%.2f' % x) # changes precision of float numbers out to 2 decimal points
```

```python
df.info() # this command tells you column names, row number, if it's null or not, and data types
```

```python
df.describe() # provides a high level overview
```

```python
df.isnull().sum() # all the columns and how many values were missing from those columns
```

```python
df.nunique() # shows number of unique values by column
```

```python
df.sort_values(by="World Population Percentage", ascending=False).head(10)
```

```python
df.corr(numeric_only=True) # shows correlation between columns
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.rcParams['figure.figsize'] = (20,7)
plt.show()
```

```python
# are there certain continents that have grown faster than others and in which ways
df.groupby('Continent').mean(numeric_only=True).sort_values(by="2022 Population", ascending=False)
```

```python
df[df['Continent'].str.contains('Oceania')] # check to see which countries are in Oceania
```

```python
df2 = df.groupby('Continent').mean(numeric_only=True).sort_values(by="2022 Population", ascending=False)
df2.plot()
```

```python
# find all names of columns

df.columns[5:13]
```

```python
# use the above command to capture only the population columns and assign a new df2
# [::-1] reverses the range

df2 = df.groupby('Continent')[df.columns[5:13][::-1]].mean(numeric_only=True).sort_values(by="2022 Population", ascending=False)
df2
```

```python
df3 = df2.transpose()
df3
```

```python
df3.plot()
```

```python
df.boxplot()

# circles represent outliers, green boxes are the averages, and the line above is the upper range
```
