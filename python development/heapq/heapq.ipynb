```python
import heapq

nums = [5, 1, 8, 3]
heapq.heapify(nums)
print(nums)  # [1, 3, 8, 5] â€“ smallest value is at index 0
```

```python
heapq.heappush(nums, 2)
print(nums)
```

```python
smallest = heapq.heappop(nums)
print(smallest)
print(nums)
```

```python
heapq.heappushpop(nums, 0)  # Push 0, then pop and return the smallest
# More efficient than doing heappush() followed by heappop()
```

```python
nums = [5, 1, 8, 3, 2, 10]
print(heapq.nsmallest(3, nums))
print(heapq.nlargest(2, nums))
```

```python
queue = []
heapq.heappush(queue, (2, "Do laundry"))
heapq.heappush(queue, (1, "Finish homework"))
heapq.heappush(queue, (3, "Play games"))

while queue:
    priority, task = heapq.heappop(queue)
    print(f"{task} (priority {priority})")
```
