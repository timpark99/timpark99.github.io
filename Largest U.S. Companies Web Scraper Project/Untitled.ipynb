```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://www.owgr.com/current-world-ranking'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')

# only one table so we can use find
table = soup.find('table')

# this finds all the headers
headers = table.find_all('th')

# list comprehension gets headers into list of text
headers_text = [header.text.strip() for header in headers]

# using pandas we create our dataframe
df = pd.DataFrame(columns = headers_text)

column_data = table.find_all('tr')

for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    
    length = len(df)
    df.loc[length] = individual_row_data
    
div = soup.find_all('script')
headers_text = [d.text.strip() for d in div]
headers_text
```
